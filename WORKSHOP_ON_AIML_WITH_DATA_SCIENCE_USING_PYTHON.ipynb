# DAY 02:- Introduction to Machine Learning (ML)

# Definition:
Use algorithms to create ML models that learn patterns in data, enabling them to make decisions and predictions.

# Types of Machine Learning:

# 1. Supervised Learning:
Involves labeled datasets where the model learns from input-output pairs.

# Common algorithms:
Linear Regression, Logistic Regression, Support Vector Machines, Decision Trees.

# 2. Unsupervised Learning:

Works with unlabeled data to find hidden patterns or intrinsic structures.

# Common algorithms:
 K-means Clustering, Hierarchical Clustering, Principal Component Analysis.

# 3. Reinforcement Learning:
 A type of learning where an agent learns by interacting with the environment to maximize cumulative rewards.

Techniques include:-

    Q-learning and Deep Q-Networks.
    
# Linear vs Logistic Regression

# Linear Regression

Theory:

Represents a linear relationship between independent and dependent variables. A straight line (best fit curve) is drawn to minimize the distance between the line and the data points.

    Implementation:
   Can be implemented using libraries like Scikit-learn in Python.


# Logistic Regression

Theory:

Used for binary classification tasks.

*Models the probability of a binary outcome using the logistic function, producing an S-shaped curve.*
    
    Implementation:

Suitable for problems where the outcome is categorical (e.g., yes/no, success/failure).


# Advanced Algorithms

# 1. Neural Networks:

Composed of interconnected nodes (neurons) that mimic the human brain.
Used for complex tasks such as image recognition, natural language processing, etc.

# 2. Generative Networks:

Models that can generate new data instances that resemble the training data.
Examples include :-

    Generative Adversarial Networks (GANs) and Variational Autoencoders (VAEs).
