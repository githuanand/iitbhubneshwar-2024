```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

## Data loading

```python
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
iris=sns.load_dataset('iris')
```

## Heatmap

```python
tc=tips.corr(numeric_only=True)
sns.heatmap(tc,annot=True,cmap='coolwarm')
#annot make the values visible
#cmap is cool warm
```

```python
fpt=flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fpt,cmap='magma',linecolor='white',lw='0.5')
```

## Clustermaps

```python
sns.clustermap(fpt,cmap='magma',lw=2,standard_scale=1,annot=True)
#standard scale makes scale used to a given standard
#cluster the data with similarity
```

## Pairgrid

```python
pg=sns.PairGrid(iris)
pg.map(sns.scatterplot)
```

```python
pg=sns.PairGrid(iris)
pg.map_diag(sns.histplot,color='violet')
#only diagonal elements
pg.map_upper(sns.scatterplot,c='red')
#only upper triangle
pg.map_lower(sns.kdeplot,fill=True,cmap='viridis')
#only lower triangle
```

## facet grid

```python
g=sns.FacetGrid(tips,col='time',row='smoker')
g.map(sns.histplot,'total_bill')
#used to make a customised grid
#divides th grid into all possible combination of given columns
```

## Regression plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
#makes a linear regression model
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker',markers=['o','v'],scatter_kws={'s':20})
#you can specify hue and markers
#scatter_kws is used to pass in plt type arguments reference the documentation
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='smoker',row='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.8,height=8)
#aspect and height can be used to make to change shape ans size
```

## styling and colors

```python
sns.set_style('white')#white backgroud
sns.countplot(x='sex',data=tips,palette='viridis')
```

```python
sns.set_style('ticks')#gives you ticks on the edges
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')#gives you ticks on the edges
sns.countplot(x='sex',data=tips)
sns.despine()
#removes the top and right by default
```

```python
#you can pass additional arguments
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

```python
plt.figure(figsize=(12,6),dpi=500)
sns.countplot(x='sex',data=tips)
#plt can be called to overide size taken by sns
```
