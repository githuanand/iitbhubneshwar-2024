```python
import seaborn as sns
import numpy as np
sns.set(style='darkgrid',palette='deep')
```

## loading a data set

```python
tips=sns.load_dataset('tips')#inbuilt data set in sns 
#uses pandas dataframe
tips.info()
```

```python
tips.head()
```

## displots

```python
sns.distplot(tips['total_bill'],bins=40)
#distribution plot used for single variable
#bins give you number of bars
'''dist plot is getting replaced by hist plot and displot'''
```

```python
sns.displot(tips['total_bill'],kde=True,bins=40)#kde=kernel distribution estimate
```

## joint plots

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex',cmap='viridis')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',fill=True,cmap='viridis')
```

## pairplots

```python
#makes a joint plot for every combination of numerical value
sns.pairplot(tips,hue='sex',diag_kind='hist',hue_order=['Female','Male'],diag_kws={"alpha":1})#hue is used for integration categorical value
```

## rugplots

```python
sns.rugplot(tips['total_bill'])#draws a line for every point
#rugplots are used to make a kde
#kde is made by sumation of each normal dsitribution on each rugplot dash 
```

## Barplots

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#gives you bar plot between average total bill and male and female
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
#estimator allows you to make and insert oyur own functions
```

## Countplots

```python
sns.countplot(x='sex',data=tips)#only uses count as an estimator
```

## Boxplots

```python
sns.boxplot(x='day',y='total_bill',data=tips)
#makes a box plot 
#each line is called quartiles and dots are outliers
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

## Violin plots

```python
sns.violinplot(x='day',y='total_bill',data=tips)
#shows the kernel density dsitribution if you split it into half
#very smiliar to box
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
#same as boxplot
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
#split allows you to split the violin plot into two and directly compare
```

## Strip plots

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
#jitter adds a little noise to take care of overlapping parts
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
#dodge is used to make the plots separate
```

## swarm plots

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
#its a combination of violin and strip plot
#not recommended for large data sets
#NOT TO BE USED
```

## Cat plots

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```
